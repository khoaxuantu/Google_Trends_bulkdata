## Import ribraries

```python
import time
import pandas as pd
import pytrends
```

## Connect to Google

```python
from pytrends.request import TrendReq
pytrends = TrendReq(hl="en-US", tz=0, retries=10)
```

## Get keywords

```python
keywords = pd.read_excel("keywords.xlsx", usecols="A")
keywords = keywords.sort_values(by=["Key"])
keys = []
for i in keywords.index:
    keys.append(keywords["Key"][i])
print(keys)
```

## Build a query loops

```python
delay = 6
print("Numbers of queries to do: ", len(keywords))

# Prepare containers
errors_list = []
count = 1
trends = {}

# Start loop
for key in keys:
    try:
        time.sleep(delay)
        pytrends.build_payload([key], timeframe="today 12-m", geo="", gprop="")
        trends[key] = pytrends.interest_over_time()[key]
        print(key + f" - Success at loop {count}")
    
    except:
        print(key + f" - Error at loop {count}")
        errors_list.append([key])
    
    count += 1

print(f"\nDone - {len(errors_list)} errors left")
```

## Tackle errors
Redo searches with errors

```python
loop = 2

if len(errors_list) > 0:
    print("\nStart errors:")

    for t in range(loop):
        if len(errors_list) > 0:
            print(f"Loop {t}: {len(errors_list)} errors left")

            for keyword in errors_list:
                try:
                    time.sleep(delay)
                    pytrends.build_payload([keyword[0]], timeframe="today 12-m", geo="", gprop="")
                    trends[keyword[0]] = pytrends.interest_over_time()[keyword[0]]
                    print(f"{keyword[0]} - Success")
                except:
                    print(f"{keyword[0]} - Error")
                    if (t == loop-1):
                        trends[keyword[0]] = pd.DataFrame()
                    pass
        else:
            break

print(f"\n Done - {len(errors_list)} errors left")
```

## Save to CSV file

```python
dict_of_trends = {"global": pd.DataFrame(trends)}
data_df = pd.concat(dict_of_trends, axis=1)
data_df.to_csv("output1.csv", index=False)
data_df.head()
```
